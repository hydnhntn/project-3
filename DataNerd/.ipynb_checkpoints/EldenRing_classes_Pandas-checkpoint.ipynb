```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of page to be scraped
url = 'https://www.pcgamer.com/all-elden-ring-classes-origins/'

# retrieve page with the reqests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html5lib')
```

```python
class_table = soup.find('table')
ths = class_table.find_all('th')
keys = []
data_dict = {}
for th in ths:
    keys.append(th.text.strip())
for key in keys:
    data_dict[key] = []
print(data_dict)
```

```python
# URL of page to be scraped
url = 'https://www.pcgamer.com/all-elden-ring-classes-origins/'

# retrieve page with the reqests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html5lib'
soup = BeautifulSoup(response.text, 'html5lib')
tbody = soup.find('tbody')
trs = tbody.find_all('tr')
for tr in trs:
    data = []
    tds = tr.find_all('td')
    for td in tds:
        data.append(td.text)
    for n in list(range(len(keys))):
        data_dict[keys[n]].append(data[n])
df = pd.DataFrame(data_dict)
df
```
