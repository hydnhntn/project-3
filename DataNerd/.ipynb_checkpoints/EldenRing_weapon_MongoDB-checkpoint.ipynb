```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of page to be scraped
url = 'https://eldenring.wiki.fextralife.com/Greatswords'

# retrieve page with the reqests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html5lib')
```

```python
thead = soup.find('thead')
ths = thead.find_all('th')
keys = []
data_dict = {}
for th in ths:
    keys.append(th.text.strip())
data_dict['Type'] = []
for key in keys:
    data_dict[key] = []
print(data_dict)
```

```python
weapons = ['Greatswords','Axes','Colossal+Swords','Colossal+Weapons','Curved+Greatswords','Curved+Swords]
for weapon in weapons:
    # URL of page to be scraped
    url = f'https://eldenring.wiki.fextralife.com/{weapon}'

    # retrieve page with the reqests module
    response = requests.get(url)
    # Create BeautifulSoup object; parse with 'html5lib'
    soup = BeautifulSoup(response.text, 'html5lib')
    tbody = soup.find('tbody')
    trs = tbody.find_all('tr')
    for tr in trs:
        data = []
        tds = tr.find_all('td')
        for td in tds:
            data.append(td.text)
        for n in list(range(len(keys))):
            data_dict[keys[n]].append(data[n])
        data_dict['Type'].append(weapon.replace('+',' '))
df = pd.DataFrame(data_dict)
df.sort_values(by=['Phy'], inplace=True, ascending=False)
df.head()
```

```python
atrs = ['Str','Dex','Int','Fai','Arc']
for atr in atrs:
    df[[f'{atr} Req',f'{atr} Scaling']] = df[atr].str.split(expand=True)
    df.pop(f'{atr}')
df[['Phy Base','Phy Modifier']] = df.Phy.str.split(expand=True).astype(int)
df.pop('Phy')
df.sort_values(by=['Phy Base'], inplace=True, ascending=False)
```

```python
pd.concat([df.pop(x) for x in ['Mag', 'Fire','Skill','Ligh','Holy']], axis=1)
df
```
